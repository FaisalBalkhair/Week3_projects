### We use the Date Fruit Dataset to build a ML model that predicts the type of date fruit based on numerical features

```python
import pandas as pd

file_path = inner_path + '/Date_Fruit_Datasets.xlsx'

df = pd.read_excel(file_path)
df.head()
```

### We classify the Features and the target

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

### We split the dataset into 80% Train, 20% Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.2, shuffle=True)
```

### We will train and compare three different models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC


logistic_model = LogisticRegression(max_iter=2000)
rForest_model = RandomForestClassifier(random_state=42)
svc_model = SVC()
```

### We will split the data to 5 parts, 4 to train and 1 to validate

```python
from sklearn.model_selection import cross_val_score

models = {
    'Logistic Regression' : logistic_model,
    'RandomForestClassifier' : rForest_model,
    'SVC' : svc_model

}

for name, model in models.items():
    scores = cross_val_score(model, X_train,y_train, cv=5)
    print(name, scores.mean())
```

```python
rForest_model.fit(X_train,y_train)
score = rForest_model.score(X_test,y_test)
score
```

### Now I will try to improve the Random Forest Model

```python
random_forset_improved = RandomForestClassifier(
    n_estimators=300,
    random_state=42
)
random_forset_improved.fit(X_train,y_train)

improved_score = random_forset_improved.score(X_test,y_test)
improved_score
```

### We tried to improve the model by tuning its hyperparameter(such as the number of trees). The final score did not change, thats mean the baseline model was already doing well.

## Count Confusion Matrix:
 
 ! Took it ready from penguins example


```python
from sklearn.metrics import confusion_matrix
import plotly.figure_factory as ff
import numpy as np

def evaluate_model(random_forset_improved, name):
    """Generate and display a confusion matrix heatmap for a given model."""
    y_pred = random_forset_improved.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    
    # Create labels for the confusion matrix
    labels = ['DOKOL', 'SAFAVI', 'ROTANA', 'DEGLET', 'SOGAY', 'IRAQI', 'BERHI']
    
    # Create annotated heatmap using plotly
    fig = ff.create_annotated_heatmap(
        z=cm,
        x=labels,
        y=labels,
        colorscale='Blues',
        showscale=True
    )
    
    fig.update_layout(
        title=f'{name} Confusion Matrix',
        xaxis_title='Predicted',
        yaxis_title='Actual',
        width=500,
        height=400
    )
    # Reverse y-axis to match sklearn convention
    fig.update_yaxes(autorange='reversed')
    fig.show()
    
# Evaluate model
evaluate_model(random_forset_improved, "Random Forest")
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score

def get_metrices(model, name):


    y_pred = model.predict(X_test)

    return{
        'Model' : name,
        'Accuracy' : accuracy_score(y_test, y_pred),
        'Precision' : precision_score(y_test, y_pred, average='weighted'),
        'Recall' : recall_score(y_test, y_pred, average='weighted'),
    }

metrics = [
    get_metrices(random_forset_improved, "Random Forest"),
]

# Create a summary DataFrame for easy comparison
df_metrics = pd.DataFrame(metrics)
df_metrics.set_index("Model", inplace=True)
df_metrics.round(4)
```
